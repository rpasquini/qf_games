```python
import sympy as sym
from sympy import diff
import matplotlib.pyplot as plt
from sympy import symbols
from sympy.plotting import plot
plt.rcParams['figure.figsize'] = 8, 5
```

```python
def OptimalAllocationGraph(U1A, U1B, U2A, U2B, D):
  # Graphs the social optimal allocation
  x = symbols('x')

  U1Aprime=diff(U1A,x)
  U1Bprime=diff(U1B,x)
  U2Aprime=diff(U2A,x)
  U2Bprime=diff(U2B,x)

  SMUA=U1Aprime+U2Aprime
  SMUB=U1Bprime+U2Bprime

  SMUBforgraph=SMUB.subs(x,D-x)
  SMUBforgraph

  p1 = plot(SMUA, (x,0,D),  xlim=[0,D], ylim=[0,20], show=False)
  p2 = plot(SMUBforgraph, (x,0,D), xlim=[0,D], ylim=[0,20], show=False)
  p1.append(p2[0])
  p1
  p1[1].line_color = 'red'
  p1.show()

```

```python
U1A=4*(x**0.5)
U1B=2*(x**0.5)
U2A=2*(x**0.5)
U2B=4*(x**0.5)
D=18
OptimalAllocationGraph(U1A, U1B, U2A, U2B, D)
```
